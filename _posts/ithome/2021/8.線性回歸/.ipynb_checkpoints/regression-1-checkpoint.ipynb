```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
# generate random data-set
np.random.seed(0)
noise = np.random.rand(100)
X = np.linspace(0,1, num=100)
y = np.power(X, 1)*.5+0.1*noise
#y=x^2+ab  Target function  a=0.2

# plot
plt.scatter(X,y,s=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 切割測試集與驗證集

```python
from sklearn.model_selection import train_test_split
X=X.reshape(-1,1)
y=y.reshape(-1,1)
X_test  = X
y_test = y
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3, random_state=42)
# mg_series=np.append(trnLbls,chkLbls)
```

## ANFIS Model

```python
import torch
import torch.nn.functional as F
```

```python
class Net(torch.nn.Module):
    def __init__(self, n_inputs, n_rules):
        super(Net, self).__init__()
        self.n = n_inputs
        self.m = n_rules
#         mu_weight=torch.from_numpy(np.random.uniform(0,1,self.n*self.m)).double()
        self.mu = torch.nn.Parameter(torch.as_tensor(torch.empty(self.n*self.m).normal_(mean=0,std=1)))
        self.sigma=torch.nn.Parameter(torch.as_tensor(torch.empty(self.n*self.m).normal_(mean=0,std=1)))
        self.y=torch.nn.Parameter(torch.as_tensor(torch.unsqueeze(torch.empty(self.m).normal_(mean=0,std=1), dim=0)),requires_grad=False)
        self.b=(torch.empty(self.n*self.m).normal_(mean=0,std=1).reshape(-1,self.m)) # bias
        
    def forward(self, x):
        self.inputs=torch.tensor(x)
        self.rul=torch.prod(torch.reshape(torch.exp(-0.5 * torch.pow(self.inputs.repeat((1, self.m))-self.mu,2)/torch.pow(self.sigma,2)),(-1, self.m, self.n)), axis=2)
        self.num = torch.sum(torch.mul(self.rul, self.y), axis=1)
#         self.num = torch.sum(torch.mul(self.rul, (torch.mul(self.inputs,self.y)+self.b)), axis=1) # function
        self.den = torch.clamp(torch.sum(self.rul, axis=1), 1e-12, 1e12)
        self.out = torch.div(self.num,self.den)
        return self.out
```

```python
torch.manual_seed(28) #14 2 19 256
net = Net(n_inputs=1, n_rules=64)     # define the network
# optimizer = torch.optim.SGD(net.parameters(), lr=0.2)
optimizer = torch.optim.Adam(net.parameters(), lr=0.001, betas=(0.9, 0.99))
loss_func = torch.nn.MSELoss()  # this is for regression mean squared loss
trn_costs = []
val_costs = []
for epoch in range(2000):
    trn_pred = net(X_train)     # input x and predict based on x
    trn_loss = loss_func(trn_pred, torch.tensor(y_train[:,0]))     # must be (1. nn output, 2. target)
    optimizer.zero_grad()   # clear gradients for next train
    trn_loss.backward()         # backpropagation, compute gradients
    optimizer.step()        # apply gradients
    val_pred = net(X_val)
    val_loss = loss_func(val_pred, torch.tensor(y_val[:,0]))
    if epoch%100==0:
        print("Train cost after epoch %i: %f" % (epoch, trn_loss),"Validation loss: %f" % val_loss)
        # plot
#         plt.figure(1)
#         pred=net(X).detach().numpy()
#         plt.scatter(X,y,s=10)
#         plt.scatter(X,pred,s=10)
#         plt.xlabel('x')
#         plt.ylabel('y')
#         plt.text(0, 0.4, 'Loss=%.3f' % trn_loss, fontdict={'size': 15, 'color':  'red'})
#         plt.savefig('./screenshot/pic'+str(epoch)+".png")
#         plt.show()
    trn_costs.append(trn_loss.detach().numpy())
    val_costs.append(val_loss.detach().numpy())
```

```python
pred=net(X).detach().numpy()

plt.figure(figsize=(20,5))
plt.figure(1)
plt.plot(y)
plt.plot(pred)
```
